```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
dataframe=pd.read_csv("cardata.csv")

dataframe.head()
```

```python
#assigning labels
df_label=dataframe["acceptability"]
df_label.head()
```

```python
#assigning features
df_features=dataframe.drop('acceptability',1)

#replacing low-1,med-2,high-3,vhigh-4 from column cost
df_features["cost"].replace("low",1,inplace=True)
df_features["cost"].replace("med",2,inplace=True)
df_features["cost"].replace("high",3,inplace=True)
df_features["cost"].replace("vhigh",4,inplace=True)

#replacing low-1,med-2,high-3,vhigh-4 from column maintain_cost
df_features["maintain_cost"].replace("low",1,inplace=True)
df_features["maintain_cost"].replace("med",2,inplace=True)
df_features["maintain_cost"].replace("high",3,inplace=True)
df_features["maintain_cost"].replace("vhigh",4,inplace=True)

#replacing 5more-5 from column doors
df_features["doors"].replace("5more",5,inplace=True)

#replacing more-5 from column person
df_features["person"].replace("more",5,inplace=True)


#replacing small-1,med-2,big-3 from column lug_boot
df_features["lug_boot"].replace("small",1,inplace=True)
df_features["lug_boot"].replace("med",2,inplace=True)
df_features["lug_boot"].replace("big",3,inplace=True)

#replacing low-1,med-2,high-3 from column safety
df_features["safety"].replace("low",1,inplace=True)
df_features["safety"].replace("med",2,inplace=True)
df_features["safety"].replace("high",3,inplace=True)

df_features.head(150)
```

```python
label=[]
for lab in df_label:
    if lab == "unacc":
        label.append([1,0,0,0])
    elif lab == "acc":
        label.append([0,1,0,0])
    elif lab == "good":
        label.append([0,0,1,0])
    else:
        label.append([0,0,0,1])   
```

```python
#converting values of features & labels into array datatype
dataArr=np.array(df_features)
labelArr=np.array(label)
```

```python
x_train,x_test,y_train,y_test=train_test_split(dataArr,labelArr,test_size=0.2,random_state=42)
x_train.shape
x_test.shape
```

```python
model = Sequential()
model.add(Dense(500, input_dim=6, activation='sigmoid'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(4, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=1000, batch_size=70, validation_data=(x_test,y_test))

```

```python
#testing model

test1=np.array([[4,4,2,2,1,1]]) #belongs to unacceptable
test2=np.array([[4,3,3,5,1,3]]) #belongs to acceptable
test3=np.array([[1,1,5,5,3,2]]) #belongs to good
test4=np.array([[1,1,5,5,3,3]]) #belongs to vgood
result=model.predict_classes(test4)
print (result)
if result==0:
    print("unacceptable")
elif result==1:
    print("acceptable")
elif result==2:
    print("good")
else:
    print("very good")
    
```
